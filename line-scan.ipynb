```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.ndimage.filters import gaussian_filter
```

```python
nx, ny = 50, 50
image = np.ones((nx, ny))
image[:,ny//2:] = 0
image[nx//2:,:ny//2] = -1
image = gaussian_filter(image, 5)

plt.imshow(image, cmap='jet')
plt.colorbar()
```

```python
vline = image[:,10]

x = np.arange(len(vline))
y = vline
plt.plot(x, y, 'k-')
left, right = (y[23], y[26])
plt.scatter(x[np.array([23,26])], (left, right))
print (r'Delta = {}'.format(right - left))
```

```python
def line_scan(line, step=1):
    assert len(line) > step
    result = np.zeros((len(line) - step - 1, 2))
    for i in np.arange(len(line)-step-1):
        left, right = line[i], line[i+step]
        result[i] = (i + step//2, right - left)
    return result
```

```python
x, y = line_scan(vline, step=4).T
plt.plot(x, y, 'k-')
```
